```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Pivot_ZAWZI_Ranked-Table 1

d1 = pd.read_csv('./data/Pivot_ZAWZI_Ranked-Table 1.csv', sep=';', decimal =',', skiprows=8, header=1)
```

```python
d1[0:12]
```

```python
d1[0:12].loc[:,:'G']
```

```python
t1 = d1.iloc[0:20,10:40]
t1 = t1.transpose()
t1
```

```python
plt.figure(figsize=(20,20))
plt.plot(t1);
```

```python
scale = np.absolute(np.sum(t1))
scale = scale / np.amax(scale, axis = 0)
scale
```

```python
# De volgende grafiek is zinloos.
# We stapelen log waardes waarbij kleine waardes een hogere concentratie betekenen.

# size is in inches w x h.
plt.figure(figsize = (20,20))

# Displacement for each bar
offset = np.zeros(len(t1[0]))

for i in range(15):
    plt.bar(t1.index, t1[i], bottom = offset)
    # Test for non overlapping when multiplying
    offset += t1[i] # * 1.01
```
