```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop=pd.read_csv('population.csv', header=[0,1,2,3,4])
svi=pd.read_csv('svi.csv')
```

```python
pop.head()
```

```python
pop.columns.names = ['Domain', 'Phylum','Class','Genera','freq']
```

```python
# pop eerst filteren op freq>25 
popdatum=pop.xs('Unnamed: 0_level_4',level='freq', axis=1, drop_level=False) 
pop32=pop.xs('32',level='freq', axis=1, drop_level=False) 
pop31=pop.xs('31',level='freq', axis=1, drop_level=False) 
pop30=pop.xs('30',level='freq', axis=1, drop_level=False) 
pop29=pop.xs('29',level='freq', axis=1, drop_level=False) 
pop28=pop.xs('28',level='freq', axis=1, drop_level=False) 
pop27=pop.xs('27',level='freq', axis=1, drop_level=False) 
pop26=pop.xs('26',level='freq', axis=1, drop_level=False) 
```

```python
from functools import reduce
data_frames = [popdatum,pop32, pop31, pop30,pop29,pop28,pop27,pop26]
pop= reduce(lambda  left,right: pd.merge(left,right,left_index=True,right_index=True,
                                            how='outer'), data_frames)

pop.head()
```

```python
# multi index kolommen plat slaan
pop.columns = pop.columns.map(' '.join).str.strip()
```

```python
pop.head()
```

```python
svi.rename(columns = {'Unnamed: 0':'datum'}, inplace = True)
pop.rename(columns = {'Unnamed: 0_level_0 Unnamed: 0_level_1 Unnamed: 0_level_2 Unnamed: 0_level_3 Unnamed: 0_level_4':'datum'}, inplace = True)
```

```python
df=svi.merge(pop, on='datum')
```

```python
df.head()
```

```python
plt.plot(df.datum,df['SVI 30'])
#plt.plot(df.datum,df['Eukaryota SAR Alveolata Cryptocaryon 32'])
plt.title('SVI_30 verloop')
plt.xlabel('datum')
plt.ylabel('SVI_30')
plt.show()
```

```python
plt.plot(df.datum,df['SVI 30'])
#plt.plot(df.datum,df['Eukaryota SAR Alveolata Cryptocaryon 32'])
plt.title('SVI_30 verloop')
plt.xlabel('datum')
plt.ylabel('SVI_30')
plt.show()
```

```python
#piek=SVI_30>80
import numpy as np
threshold = 80.0
x = df.datum

# split it up
above_threshold = np.maximum(df['SVI 30'] - threshold, 0)
below_threshold = np.minimum(df['SVI 30'], threshold)

# and plot it
fig, ax = plt.subplots()
ax.bar(x, below_threshold, 0.35, color="g")
ax.bar(x, above_threshold, 0.35, color="r",
        bottom=below_threshold)

# horizontal line indicating the threshold
ax.plot([0., 32], [threshold, threshold], "k--")
```

```python
#selecteer alleen datums met SVI>80 of SVI<30
df2=df.loc[(df['SVI 30'] > 80) | (df['SVI 30'] < 30)]
# en wat overblijft is alles tussen 30<SVI<80
df3=df.loc[(df['SVI 30'] <= 80) & (df['SVI 30'] >= 30)]
```

```python
# vind correlaties in df2 en vergelijk met correlaties in df
corr_df2=df2.corr()
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corr_df2, cmap=cmap, center=0)
```

```python
# sterkste correlaties in totale set
c = df.corr().abs()
s = c.unstack()
so = s.sort_values(kind="quicksort")
so[-(df.shape[1]+10):-df.shape[1]]
```

```python
# sterkste correlaties in set met alleen pieken en dalen
c2 = df2.corr().abs()
s2 = c2.unstack()
so2 = s2.sort_values(kind="quicksort")
so2[-(df2.shape[1]+10):-df2.shape[1]]
```

```python
# sterkste correlaties in set zonder pieken en dalen
c3 = df3.corr().abs()
s3 = c3.unstack()
so3 = s3.sort_values(kind="quicksort")
so3[-(df3.shape[1]+10):-df3.shape[1]]
```

```python
# vind grootste verschillen tussen c3 en c2
verschil=c3.subtract(c2)
verschil.unstack().sort_values(kind="quicksort")
```
